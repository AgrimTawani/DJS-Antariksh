Populating a csv file using pandas


```python
import random
company_names = [
    "Google", "Microsoft", "Apple", "Amazon", "Facebook", "Tesla", "Netflix", 
    "Adobe", "Intel", "IBM", "Oracle", "Cisco", "HP", "Dell", "NVIDIA", 
    "Salesforce", "Qualcomm", "Uber", "Lyft", "Airbnb"
]

companies = [random.choice(company_names) for _ in range(200)]

df = pd.DataFrame({'company_name': companies})
df.to_csv('companies.csv', index=False)
```

```python
data2 = pd.read_csv('companies.csv')
data2
```

<h1> Matplotlib </h1>

```python
import matplotlib.pyplot as plt 
import pandas as pd

df = pd.read_csv('short_placements.csv')

x = df['cgpa']
y = df['package']

plt.plot(x,y)
```

```python
plt.xlabel('cgpa', fontsize=18)
plt.ylabel('package', fontsize=16)

plt.bar(x,y)
```

```python
plt.pie(y, labels=x, radius=1.2)
```

```python
plt.xlabel('cgpa', fontsize=18)
plt.ylabel('package', fontsize=16)

plt.scatter(x,y)

```

```python
import numpy as np
x = np.linspace(0, 10*np.pi, 1000)
y = np.sin(x)
fig, ax = plt.subplots()
ax.set_title("This is a Sine wave")
ax.plot(x,y)
```
