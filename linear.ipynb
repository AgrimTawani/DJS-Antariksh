# Linear Regression using Scikit-learn

```python
import pandas as pd
import matplotlib.pyplot as plt  
from sklearn.model_selection import train_test_split
diamond_data = pd.read_csv('Diamond.csv')
X = diamond_data['carat']
y = diamond_data['price']
my_df = diamond_data[['carat', 'price']]
my_df
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(X,y,alpha=0.3)
```

Spliting the dataset 30% for training the model and rest for testing our model

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
print(X_train)
print(" ")
print(y_train)
```

- The above data is one dimensional 
- sklearn only takes two dimensional data
- hence converting this data to two dimensional

```python
import numpy as np  
X_train = np.array(X_train).reshape(-1, 1)
X_test= np.array(X_test).reshape(-1, 1)
y_train = np.array(y_train).reshape(-1, 1)
y_test= np.array(y_test).reshape(-1, 1)
```

Applying linear regression method


```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
# m = x coefficient , c = y intercept => Y = mX + c
m, c = lr.coef_, lr.intercept_ 
m, c
```

```python
Y_predicted = lr.predict(X_test)

plt.scatter(X_test, y_test)
plt.plot(X_test, Y_predicted, color="red")
plt.xlabel("Carat")
plt.ylabel("Prices")
```
